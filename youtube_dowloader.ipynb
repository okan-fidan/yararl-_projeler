```python
from pytube.request import stream
from pytube import YouTube
xx=input("indirmek istediğiniz video urlsini giriniz: ")
ss=input("video açıklamasını görmek istermisiniz yes/no: ")
yt=YouTube(xx)
if ss=="yes":
  print(f"Video Başlığı: {yt.title}")
  print(f"Video Sahibi: {yt.author}")
  print(f"Thumbnail Resmi: {yt.thumbnail_url}")
  print(f"Video Uzunluğu: {yt.length}")
  print(f"Video Rating: {yt.rating}")
  print(f"İzlenme Sayısı: {yt.views}")
  print("*"*30)
  print(f"Video Açıklaması: {yt.description}")
  print("*"*30)
elif ss=="no":
  print(f"Video Başlığı: {yt.title}")
  print(f"Video Sahibi: {yt.author}")
  print(f"Thumbnail Resmi: {yt.thumbnail_url}")
  print(f"Video Uzunluğu: {yt.length}")
  print(f"Video Rating: {yt.rating}")
  print(f"İzlenme Sayısı: {yt.views}")
  print("*"*30)

for i in yt.streams.filter(progressive=True):
  print(i)
onay=input(" indirmek için onay veriyor musunuz yes/no")
if onay=="yes":
  streamm=yt.streams.filter(progressive=True).first()
  streamm.download()
else :
  print("işlem iptal edildi")
```
